Importing the pandas library

```python
import pandas as pd 
```

Import the json Imdb Dataset

```python
# Your code goes here
data= pd.read_json('movies_dataset.json')
data.head()
```

Answer to the following Analytical questions: 

Question 1: Which country has gained the highest average IMdb Score according to our data? <br>
Use a bar chart to visualize your results

```python
data1= data.groupby(data['country'])['imdb_score'].mean()
data1.head(10)
```

```python
import matplotlib.pyplot as plt
ax=data1.plot.bar(figsize=(16,8))
plt.xlabel('Country')
plt.ylabel('IMDB Score')
plt.title('Country over IMDB score')
plt.show()
```

Question 2: How many `colored` and `Black&White` movies do we have in our dataset?

```python
# Write your code here
data2= data.groupby(data['color']).count()
data2
```

Question 3: Which movies of `Christopher Nolan` has the highest IMdb Score? <br>
Use a barchart to visualize the results

```python
data3 = data[data['director_name'] == 'Christopher Nolan']
data3
```

```python
data3_sorted = data3.sort_values(by='imdb_score', ascending=False)
data3_sorted
```

```python
import matplotlib.pyplot as plt

ax=data3_sorted.imdb_score.plot.bar(figsize=(16,8))
plt.xlabel('IMDB Score')
plt.ylabel('Movie Title')
plt.title('Christopher Nolan Movies by IMDB Score')
plt.show()
```

```python

ax=data3_sorted.plot.bar(figsize=(16,8), x='movie_title', y='imdb_score')
plt.xlabel('Movie Title')
plt.ylabel('IMDB Score')
plt.title('Christopher Nolan Movies by IMDB Score')
plt.show()
```

Question 4: Use a line chart to visualize how many movies have been produced in `USA` from `2010` to `2015`. 

```python
# Write your code here
year_data=data[(data.country=='USA') & 
          (data.title_year >2010) & 
          (data.title_year <2015)]
year_data.shape
```

```python
year_data.head()
```

```python
data3= year_data.groupby('title_year').size()
data3
```

```python
data3= year_data.groupby('title_year').size().plot()
data3
```

```python
ax=data3.plot(figsize=(16,8), x='title_year', y='country')
plt.xlabel('Year')
plt.ylabel('QTY')
plt.title('USA Movie')
plt.show()
```
